```python
result = train_country_newbies_model(
    df,
    seed=42,
    holdout_days=30,
    half_life_days=60,
    top_n_countries=15,
    target_column = 'vertical_newbies_delivery_paid',
    catboost_params={"iterations": 3000, "learning_rate": 0.05}
)
```
